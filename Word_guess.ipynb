```python
#import import random module
import random

# Make a list that contains random words to be guessed by the user.
# can also import dictionary or list english thesaurus
words = ['laptop','grocery','pencil','weather','rainbow','mountain','utensil','screen','clouds','keyboard',
         'python','gamer','mouse','speaker','cheater','dashboard','car','spot','paper','recliner','funny','suspence',
         'hunger','cake','slice','hamper','goal','distance','bed','jaggery','jaggies','jaggy','Jagiello','beautiful','because',
         'become','bed','before','begin','behavior','behind','believe','benefit','best','better']

#Randomize the list and and store a word in a variable
r_word = random.choice(words)

#the number of guesses a participant gets is fixed
n_guess = 8

#Information about game is given to player
name = input('Name: ')
print('Hello ', name )
print("Guess the word of the day. You will be given hint. Guess the word in 8 chances by guessing one letter at a time.")
print('One chance will be deducted for each wrong guess and invalid input.')
print(" ")
print("You get ", n_guess,"to guess the word")

#create an empty list to store the words guessed by the user
g_word = []
print("the length of the the letter to be guessed is :",len(r_word))

#guessed to be intialized as false
#Processing the word input by player, whether the input word is of same length
guessed =False
while guessed == False and n_guess>0:
    print("You have",n_guess,"guesses left to guess the word")
    guess = input("Guess the word.".lower())
    if len(guess) == len(r_word):
        if guess == r_word:
            g_word.append(guess)
            print("Wuhoo!! You have gussed the word. \n The word of the day is ", r_word)
        else:
            print("Nah!!", "\U0001F923", "Guess again ")
            n_guess -= 1
    else:
        print("Guess a word of ", len(r_word), 'letters')
        n_guess -=1
    
# Initializing the empty places in place of letters of 'Word of the Day'
    word_pos =' '
    if guessed == False:
        for letter in r_word:
            if letter in guess:
                word_pos = word_pos+letter
            else:
                word_pos =word_pos+'_'+' '
        print(word_pos)
        
# Output of results of the game played, whether the player won or not  
    if word_pos == r_word:
        print("hooray, you guessed the word")
        guessed= True
        break
    elif n_guess == 0:
        print(" no more guesses left, try again later")
        print("the word to be guessed is:",r_word)
```
